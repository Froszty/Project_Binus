```python
import io 
df = pd.read_csv(io.StringIO(filenya['insurance_premiums.csv'].decode('utf-8')))
print(df)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['State'], df['fatal_collisions'], label='fatal_collisions')
plt.scatter(df['State'], df['fatal_collisions_speeding'], label='fatal_collisions_speeding')
plt.scatter(df['State'], df['fatal_collisions_alc'], label='fatal_collisions_alc')
plt.scatter(df['State'], df['fatal_collisions_not_distracted'], label='fatal_collisions_not_distracted')
plt.scatter(df['State'], df['fatal_collisions_no_hist'], label='fatal_collisions_no_hist')

plt.xlabel('State')
plt.ylabel('Collisions')
plt.title('Insurance Premium of Collisions')
plt.grid(True)
plt.legend() 
```

```python
df_stacked=df.groupby(['Region']).sum().reset_index()
df_stacked
```

```python
import numpy as np
plt.figure(figsize=(10,10))

x=np.arange(len(df_stacked['Region']))
width = 0.35

plt.bar(x - width/2, df_stacked['premiums'], width, label='premiums')
plt.bar(x + width/2, df_stacked['insurance_losses'], width, label='insurance_losses')

plt.xticks(x, ['Midwest', 'Northeast', 'South', 'West'])
plt.xlabel('Region')
plt.grid(True)
plt.legend()
plt.title('Perbandingan Jumlah Premi Asuransi Pada Setiap Region')
```

```python
plt.figure(figsize=(10,10))

plt.bar(x - 0.3, df_stacked['fatal_collisions'], width = 0.1, label='fatal_collisions')
plt.bar(x - 0.2, df_stacked['fatal_collisions_speeding'], width = 0.1, label='fatal_collisions_speeding')
plt.bar(x - 0.1, df_stacked['fatal_collisions_alc'], width = 0.1, label='fatal_collisions_alc')
plt.bar(x , df_stacked['fatal_collisions_not_distracted'], width = 0.1, label='fatal_collisions_not_disctracted')
plt.bar(x + 0.1, df_stacked['fatal_collisions_no_hist'], width = 0.1, label='fatal_collisions_no_hist')

plt.xticks(x, ['Midwest', 'Northeast', 'South', 'West'])
plt.xlabel('Region')
plt.grid(True)
plt.legend
plt.title('Perbandingan Collisions Antar Regional')
```
